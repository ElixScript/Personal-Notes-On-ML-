# Random Forest Regressor

## data prep

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

## training the random forest regression model 

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10,random_state =0)
regressor.fit(x,y)
```

## predicting a result

```python
regressor.predict([[6.5]])
```

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='red')
plt.plot(x_grid,regressor.predict(x_grid),color = 'blue')
plt.title('Random Forest Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## some personal notes :

### how random forest regression work ?
this model is a further step of decision tree . it works with combine a lot of decision tree to create more accurate and stable predictions 

### what is n_estimators ?
parameter 'n_estimator' control the amount of tree (decision trees) which the random forest going to build . the n_estimator affect to stability , accuracy and overfitting risk
