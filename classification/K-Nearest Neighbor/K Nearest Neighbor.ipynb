# K Nearest Neighbor

# data prep

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(x,y,test_size = 0.25,random_state =0)
```

## feature scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## training the knn model 

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p =2)
classifier.fit(x_train,y_train)
```

## predicting a result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## predicting the test set result

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## making confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

## visualising the training set result (not detailed version)

```python
from matplotlib.colors import ListedColormap
X_set = sc.inverse_transform(x_train)
y_set = y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=5),
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=500)
)
predictions = classifier.predict(sc.transform(np.c_[X1.ravel(), X2.ravel()])).reshape(X1.shape)

plt.contourf(X1, X2, predictions, alpha=0.75, cmap=ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

colors = ListedColormap(('salmon', 'dodgerblue'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=colors(i), label=j)

plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## visualising the test set result

```python
from matplotlib.colors import ListedColormap

# Inverse transform hanya pada X_set yang diperlukan
X_set = sc.inverse_transform(x_test)
y_set = y_test

# Mengurangi resolusi meshgrid untuk mempercepat perhitungan
X1, X2 = np.meshgrid(
    np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=5),
    np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=500)
)

# Prediksi dioptimalkan dengan mengurangi transformasi array
predictions = classifier.predict(sc.transform(np.c_[X1.ravel(), X2.ravel()])).reshape(X1.shape)

# Plotting dengan contour dan scatter
plt.contourf(X1, X2, predictions, alpha=0.75, cmap=ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Loop untuk scatter plot tetap sama
colors = ListedColormap(('salmon', 'dodgerblue'))
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=colors(i), label=j)

plt.title('K-NN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
